```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv(
    'https://raw.githubusercontent.com/teeloren/freecodecamp/main/adult.data.csv')
```

```python
sal = data[data['salary'] == '>50K']
high_sal = sal['age']

plt.figure(figsize=(20,10))
plt.hist(high_sal)
plt.xlabel('age')
plt.ylabel('total')
plt.title('Age of Salary Earners Making 50K+')
# plt.grid(axis='y',color='y',lw='1')
plt.show()
```

```python

plt.figure(figsize=(20, 8), dpi=300)
plt.scatter(data['marital-status'],data['hours-per-week'])
plt.xlabel('marital-status', c='r')
plt.ylabel('hours-per-week', c='r')
plt.title('Hours Worked Per Week by Marital Status', c='r')
plt.grid(color='b', linestyle=':',linewidth='0.75')
plt.show()
```

```python
labels = ['White','Black','Asian-Pac-Islander','Amer-Indian-Eskimo','Other']
colors = ["g", "b", "c", "#f07c3a", "#6642f5"]   
explode = (0, 0.1, 0.2, 0.4,0.6)           
plt.figure(figsize=(20,10))
plt.pie(sal['race'].value_counts(), labels=labels,colors=colors, explode=explode, autopct='%1.1f%%')
plt.legend(labels=labels, bbox_to_anchor=(1, 1))
plt.title("Earners Making Over 50K by Race")
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(40,10))
labels=['Married-civ-spouse','Never-married','Divorced','Widowed','Separated','Married-spouse-absent','Married-AF-spouse']
plt.bar(sal['marital-status'], sal['age'])
plt.xlabel("Age")
plt.ylabel('Marital-status')
plt.title('Age by Marital-status')
# plt.legend(labels=labels, bbox_to_anchor=(1, 1))
# plt.grid()
plt.show()
```
